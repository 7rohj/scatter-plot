```python
import pandas as pd
import warnings
# warnings.filterwarnings(action='ignore')

import matplotlib.pyplot as plt
import numpy as np

# 엑셀 파일에서 모든 시트를 데이터프레임으로 불러오기
excel_file = '교통량.xlsx'  # 엑셀 파일 이름

# 특정 시트 이름으로 불러오기
df1 = pd.read_excel(excel_file, sheet_name='군집1')  # 시트 이름으로 지정
df2 = pd.read_excel(excel_file, sheet_name='군집2') 
df3 = pd.read_excel(excel_file, sheet_name='군집3') 
df4 = pd.read_excel(excel_file, sheet_name='군집4') 
df5 = pd.read_excel(excel_file, sheet_name='군집5') 
df6 = pd.read_excel(excel_file, sheet_name='군집6')
df6

```

```python
df_melted1 = df1.melt(id_vars='구분', var_name='hour', value_name='vehicle')
df_melted2 = df2.melt(id_vars='구분', var_name='hour', value_name='vehicle')
df_melted3 = df3.melt(id_vars='구분', var_name='hour', value_name='vehicle')
df_melted4 = df4.melt(id_vars='구분', var_name='hour', value_name='vehicle')
df_melted5 = df5.melt(id_vars='구분', var_name='hour', value_name='vehicle')
df_melted6 = df6.melt(id_vars='구분', var_name='hour', value_name='vehicle')
df_melted6
```

```python
plt.figure(figsize = (20, 5))
plt.scatter('hour', 'vehicle', data = df_melted1, color='b', label='교통량')
plt.ylabel('vehicle', fontsize=12)
plt.xlabel('hour', fontsize=12)
plt.xticks(range(1, 25, 1))
# plt.yticks(range(0, 22000, 2000))

plt.savefig('graph1.png', format='png', dpi=300)

plt.show()
```
